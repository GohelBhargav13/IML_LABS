```python
# Machine learning Lab-2 (Practical - 2 and Practical -3)

# Importing needed libreries

import numpy as np # To perform numeric operation
import random as re

# Problem - 1 - To Convert a list of a numeric values into a one-dimensional Numpy array

list = [2,3,4,5,6]

print("Before list:",list)

# Convert list into a numpy array

arr = np.array(list)

print("After list: ",arr)
```

```python
# Practical-2  Problem - 2 - To create a 3 * 3 matrix with values ranging from 2 to 10

var = np.arange(2,11).reshape(3,3)
print(var)
```

```python
# Practical-2 Problem - 3 -To append values at the end of an array

print(np.append(np.array([1,2,3,4,5]), re.randint(1,2)).reshape(3,2))
```

```python
# Practical-2 Problem - 4 - To create another shape from an array without changing its data(3*2 to 2*3)

print(np.arange(12,18).reshape(3,2).reshape(2,3))
```
