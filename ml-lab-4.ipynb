```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random as re


#                                 Practical - 4
# Practical(4) - Problem - 1 - To add, subtract, multiply, divide arguments element wise

print("Element wise Addition : " , np.add(50, 50))
print("Element wise Subtraction : " , np.subtract(50, 50))
print("Element wise Multiplication : " , np.multiply(50, 50))
print("Element wise Division : " , np.divide(50, 50))

```

```python
# Practical(4) - Problem - 2 - To round elements of the array to the nearest integer

print("Actucal array : ", np.arange(10.5, 15.5), "\nRounded array : ", np.rint(np.arange(10.5, 15.5)))
```

```python
# Practical(4) - Problem - 3 - To calculate mean across dimension, in a 2D array

print("Row mean : ", np.mean(np.arange(1, 10).reshape(3, 3), axis = 1))
print("Column mean : ", np.mean(np.arange(1, 10).reshape(3, 3), axis = 0))
```

```python
# Practical(4) - Problem - 4 - To calculate the difference between neighboring elements, element wise of a given array

print(np.diff(np.array([10, 20, 25, 30, 34, 50])))
```
