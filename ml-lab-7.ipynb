```python
#Practical-(7) - Problem -3 to delete the one specific column from the DataFrame

d = {
    'Name':["Bhargav","Chhayank","Rohit","Aditya"],
    'Age' : [18,18,18,18],
    'Enrollment' : [13,17,12,32]
}

df =pd.DataFrame(d)

df.drop(['Age'], axis=1)
```

```python
#Practical-(7) - Problem -4 to write a DataFrame to CSV file using tab separator

d = {
    'Name':["Bhargav","Chhayank","Rohit","Aditya"],
    'Age' : [18,18,18,18],
    'Enrollment' : [13,17,12,32]
}

data_frame = pd.DataFrame(d, columns = ['Name', 'Age', 'Enrollment'])

data_frame.to_csv('Details1.csv', sep = '\t') # Saving as a CSV file
 
new_data_frame = pd.read_csv('Details1.csv') # loading the CSV file
 
print('\nData from Students.csv : \n', new_data_frame)
```
