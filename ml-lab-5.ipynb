```python
#                                      Practical-5
#Practical-(5)-problem  - 1- to Find the maximum and minimum value of a given flattened array

print("Actual Array: \n",np.arange(11,15).reshape(2,2), "\nMaximum : ",np.amax(np.arange(11,15).reshape(2,2)), "\nMinimum : ",np.amin(np.arange(11,15).reshape(2,2)))
```

```python
#                                      Practical-5
#Practical(5)- Problem - 2 - to compute the mean, standard deviation, and variance of a given array along the second axis

print("Actual Array :",np.arange(11,21), "\nMean : ", np.mean(np.arange(11,21)), "\n Standard deviation : ",np.std(np.arange(11,21)), "\n Variance : ",np.var(np.arange(11,21)))
```
